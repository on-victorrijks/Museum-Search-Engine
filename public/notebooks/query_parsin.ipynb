```python
import spacy
from spacy.lang.fr.examples import sentences 

nlp = spacy.load("fr_dep_news_trf")
```

```python
def analyser_requete(requete):
    doc = nlp(requete)
    
    mots_visuels = []
    mots_metadata = []
    
    for token in doc:
        if token.dep_ in ('nsubj', 'dobj', 'attr'):  # Sujets, objets, attributs (visuels)
            mots_visuels.append(token.text)
        elif token.text.lower() in ['du', 'de', 'en', 'dans']:  # Mots-clés liés aux métadonnées
            mots_metadata.append(' '.join([t.text for t in doc[token.i+1:]]))  # Extraire après 'du', 'de', etc.
    
    return mots_visuels, mots_metadata

requete = "Montre-moi des peintures avec des bateaux du XVIIIe siècle"
visuels, metadata = analyser_requete(requete)
print("Mots-clés visuels:", visuels)
print("Mots-clés metadata:", metadata)
```
